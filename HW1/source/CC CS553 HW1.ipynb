```python
import matplotlib.pyplot as plt
dns_names = []
avg_rtt_values = []
with open("network-test-latency.txt", 'r') as file:
  for line in file:
            dns_name, avg_rtt = line.split()
            dns_names.append(dns_name)
            avg_rtt_values.append(float(avg_rtt))
plt.figure(figsize=(10, 6))
plt.plot(dns_names, avg_rtt_values, marker='o', linestyle='-', color='b')
plt.xlabel('DNS Name')
plt.ylabel('Average RTT (ms)')
plt.title('Network Latency Line Chart')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
